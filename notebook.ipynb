```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

chemin_fichier = 'dataset.csv'
df = pd.read_csv(chemin_fichier)

df
```

```python
# Types de données
print(df.dtypes)

# Nombre de valeurs manquantes par colonne
print(df.isnull().sum())
```

```python
# Statistiques descriptives
print(df.describe())
```

```python
# Calcul des corrélations en excluant les colonnes non numériques
correlations = df.corr(numeric_only=True)

# Affichage des corrélations
print(correlations)

# Pour visualiser les corrélations avec seaborn
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
sns.heatmap(correlations, annot=True, fmt=".2f", cmap='coolwarm')
plt.show()

```

```python
import pandas as pd

# Nettoyage de la colonne 'publishedsince' pour ne garder que le nombre
df['publishedsince'] = df['publishedsince'].astype(str).str.extract('(\d+)').astype(int)

# Enlever les '\n' de la colonne 'carmodel' et suppression des espaces superflus
df['carmodel'] = df['carmodel'].astype(str).replace('\n', '').str.strip()

# Nettoyage de la colonne 'price' pour ne garder que les valeurs numériques
df['price'] = df['price'].astype(str).str.replace(' ', '').str.extract('(\d+)').astype(float)

# Conversion de la colonne 'année' en entier
df['année'] = df['année'].fillna(0).astype(int)  # Remplacer NaN par 0 ou une autre valeur par défaut avant la conversion

# Conversion de la colonne 'miseencirculation' en date
df['miseencirculation'] = pd.to_datetime(df['miseencirculation'], format='%d/%m/%Y', errors='coerce')

# Nettoyage de la colonne 'kilométragecompteur' pour ne garder que les valeurs numériques entières
df['kilométragecompteur'] = df['kilométragecompteur'].astype(str).str.extract('(\d+)').astype(float)

# Nettoyage de la colonne 'puissancemoteur' pour ne garder que le nombre
df['puissancemoteur'] = df['puissancemoteur'].astype(str).str.extract('(\d+)').astype(float)

# Nettoyage de la colonne 'voltagebatterie' pour ne garder que les nombres
df['voltagebatterie'] = df['voltagebatterie'].astype(str).str.extract('(\d+)').astype(float)

# Affichage des premières lignes pour vérifier le nettoyage
print(df.head())

```

```python
import numpy as np

# Remplacer les années clairement erronées par NaN
df['année'] = df['année'].apply(lambda x: np.nan if x < 1885 else x)

# Optionnellement, remplacer les NaN par l'année moyenne/médiane (si cela a du sens pour votre analyse)
median_year = df['année'].median()
df['année'].fillna(median_year, inplace=True)

# Convertir à nouveau en entier si nécessaire
df['année'] = df['année'].astype(int)
```

```python
print(sorted(df['année'].unique()))
print(df['année'].describe())
```

```python
# Statistiques descriptives pour toutes les colonnes numériques
print(df.describe())

# Pour une colonne spécifique, par exemple 'price'
print(df['price'].describe())
```

```python
# Sélection des colonnes numériques
numeric_columns = df.select_dtypes(include=['number'])

# Calcul des corrélations
correlations = numeric_columns.corr()

# Affichage des corrélations
print(correlations)

# Pour une visualisation plus intuitive des corrélations
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
sns.heatmap(correlations, annot=True, cmap='coolwarm', fmt=".2f")
plt.show()

```

```python
# Histogramme pour 'price'
df['price'].hist(bins=30)
plt.title('Distribution des prix')
plt.xlabel('Prix')
plt.ylabel('Nombre de véhicules')
plt.show()
```

```python
# Boxplot pour 'price'
sns.boxplot(x=df['price'])
plt.title('Boxplot des prix')
plt.show()

# Boxplot comparatif par catégorie, par exemple 'énergie'
sns.boxplot(x='énergie', y='price', data=df)
plt.title('Boxplot des prix par type d\'énergie')
plt.xticks(rotation=45)  # Rotation des étiquettes si nécessaire
plt.show()
```

```python
# Relation entre 'kilométragecompteur' et 'price'
sns.scatterplot(x='kilométragecompteur', y='price', data=df)
plt.title('Relation entre kilométrage et prix')
plt.xlabel('Kilométrage')
plt.ylabel('Prix')
plt.show()
```

```python
# Comptage des valeurs pour une variable catégorielle, par exemple 'énergie'
print(df['énergie'].value_counts())

# Visualisation des comptes
sns.countplot(data=df, x='énergie')
plt.title('Distribution des types d\'énergie')
plt.xticks(rotation=45)
plt.show()
```

```python
# Tendances au fil du temps, si applicable, par exemple prix moyen par année
df.groupby('année')['price'].mean().plot(kind='line')
plt.title('Prix moyen par année')
plt.xlabel('Année')
plt.ylabel('Prix moyen')
plt.show()
```
